```python
! pip install scapy
import pandas as pd
import matplotlib.pyplot as plt
import plotly.express as px
import seaborn as sns
import scipy
from scapy.all import rdpcap
import datetime
import numpy as np
from datetime import datetime
import matplotlib.pyplot as plt
import time
import copy
import statistics
from sklearn.metrics import mean_squared_error
from google.colab import drive
drive.mount('/content/drive')

from google.colab import files
```

```python
fileName = "<PATH_FILENAME>.pcap"

x=fileName.split(sep='/')
x=x[-1][0]
captureClass = int(x)            #class Normal = 0
                                 #class Cavitation = 1
                                 #class Dry Running = 2  

inicio = time.time()

packets = rdpcap(fileName)

actualSpeed = []
actualCurrent = []
actualTorque = []
actualActivePower = []
aS1 = []
aS2 = []

for i in range(len(packets)):
    if (packets[i].src+packets[i].dst == packets[0].src+packets[0].dst) and (len(packets[i].original)==60):
        b7 = str(hex(packets[i].original[len(packets[i].original)-37]))
        b7_1 = b7.split('0x')
        if len(b7_1[1]) == 1:
            b7='0'+b7_1[1]
        else:
            b7 = b7_1[1]

        b6 = str(hex(packets[i].original[len(packets[i].original)-36]))
        b6_1 = b6.split('0x')
        if len(b6_1[1]) == 1:
            b6='0'+b6_1[1]
        else:
            b6 = b6_1[1]

        b5 = str(hex(packets[i].original[len(packets[i].original)-35]))
        b5_1 = b5.split('0x')
        if len(b5_1[1]) == 1:
            b5='0'+b5_1[1]
        else:
            b5 = b5_1[1]    

        b4 = str(hex(packets[i].original[len(packets[i].original)-34]))
        b4_1 = b4.split('0x')
        if len(b4_1[1]) == 1:
            b4='0'+b4_1[1]
        else:
            b4 = b4_1[1]

        b3 = str(hex(packets[i].original[len(packets[i].original)-33]))
        b3_1 = b3.split('0x')
        if len(b3_1[1]) == 1:
            b3='0'+b3_1[1]
        else:
            b3 = b3_1[1]    

        b2 = str(hex(packets[i].original[len(packets[i].original)-32]))
        b2_1 = b2.split('0x')
        if len(b2_1[1]) == 1:
            b2='0'+b2_1[1]
        else:
            b2 = b2_1[1]    

        b1 = str(hex(packets[i].original[len(packets[i].original)-31]))
        b1_1 = b1.split('0x')
        if len(b1_1[1]) == 1:
            b1='0'+b1_1[1]
        else:
            b1 = b1_1[1]    

        b0 = str(hex(packets[i].original[len(packets[i].original)-30]))
        b0_1 = b0.split('0x')
        if len(b0_1[1]) == 1:
            b0='0'+b0_1[1]
        else:
            b0 = b0_1[1]
        aS1_support = [b6]
        aS2_support = [b7]
        aS1.append(aS1_support)
        aS2.append(aS2_support)
        actualActivePower_support = '0x'+b1+b0
        actualActivePower_support = (int(actualActivePower_support,16)/int("4000",16))
        actualActivePower.append(actualActivePower_support)  
        actualTorque_support = '0x'+b3+b2
        actualTorque_support = (int(actualTorque_support,16)/int("4000",16))
        actualTorque.append(actualTorque_support)  
        actualCurrent_support = '0x'+b5+b4
        actualCurrent_support = (int(actualCurrent_support,16)/int("4000",16))
        actualCurrent.append(actualCurrent_support)        
        actualSpeed_support = '0x'+b7+b6
        actualSpeed_support = (int(actualSpeed_support,16)/int("4000",16))
        actualSpeed.append(actualSpeed_support)
        
#Media de cada atributo por segundo de operação (1000 pacotes)

i=0
r = 1000
#Atributos de Corrente
avgCurrent = []
stdCurrent = []
maxCurrent = []
minCurrent = []
rangeCurrent = []
kurtosisCurrent = []
rmsCurrent = []
skewnessCurrent = []
entropyCurrent = []

avgSpeed = []

#Atributos de Torque
avgTorque = []
stdTorque = []
maxTorque = []
minTorque = []
rangeTorque = []
kurtosisTorque = []
rmsTorque = []
skewnessTorque = []
entropyTorque = []

#Atributos de Fator de Potencia
avgActivePower = []
stdActivePower = []
maxActivePower = []
minActivePower = []
rangeActivePower = []
kurtosisActivePower = []
rmsActivePower = []
skewnessActivePower = []
entropyActivePower = []
listClass = []
perSecond = []

for i in range(0,len(actualCurrent),r):

    mediaCurrent = '%.6f'%float(sum(actualCurrent[i:i+r])/len(actualCurrent[i:i+r]))
    avgCurrent.append(mediaCurrent)
    getKurtosisCurrent = '%.6f'%float(abs(scipy.stats.kurtosis(actualCurrent[i:i+r],bias=False)))
    kurtosisCurrent.append(getKurtosisCurrent)
    getSkewnessCurrent = '%.6f'%float(abs(scipy.stats.skew(actualCurrent[i:i+r],bias= True)))
    skewnessCurrent.append(getSkewnessCurrent)
    getEntropyCurrent = '%.6f'%float(abs(scipy.stats.entropy(actualCurrent[i:i+r])))
    entropyCurrent.append(getEntropyCurrent)
    getMaxCurrent = '%.6f'%float(max(actualCurrent[i:i+r]))
    maxCurrent.append(getMaxCurrent)
    getMinCurrent = '%.6f'%float(min(actualCurrent[i:i+r]))
    minCurrent.append(getMinCurrent)
    getRangeCurrent = '%.6f'%float(float(getMaxCurrent)-float(getMinCurrent))
    rangeCurrent.append(getRangeCurrent)
    getStdCurrent = '%.6f'%float(abs(statistics.stdev(actualCurrent[i:i+r])))
    stdCurrent.append(getStdCurrent)
    getRmsCurrent = '%.6f'%float(abs(mean_squared_error(actualCurrent[i:i+r],[0 for _ in actualCurrent[i:i+r]], squared=False)))
    rmsCurrent.append(getRmsCurrent)

    mediaTorque = '%.6f'%float(sum(actualTorque[i:i+r])/len(actualTorque[i:i+r]))
    avgTorque.append(mediaTorque) 
    getKurtosisTorque = '%.6f'%float(abs(scipy.stats.kurtosis(actualTorque[i:i+r],bias=False)))
    kurtosisTorque.append(getKurtosisTorque)
    getSkewnessTorque = '%.6f'%float(abs(scipy.stats.skew(actualTorque[i:i+r],bias= True)))
    skewnessTorque.append(getSkewnessTorque)
    getEntropyTorque = '%.6f'%float(abs(scipy.stats.entropy(actualTorque[i:i+r])))
    entropyTorque.append(getEntropyTorque)
    getMaxTorque = '%.6f'%float(max(actualTorque[i:i+r]))
    maxTorque.append(getMaxTorque)
    getMinTorque = '%.6f'%float(min(actualTorque[i:i+r]))
    minTorque.append(getMinTorque)
    getRangeTorque = '%.6f'%float(float(getMaxTorque)-float(getMinTorque))
    rangeTorque.append(getRangeTorque)
    getStdTorque = '%.6f'%float(abs(statistics.stdev(actualTorque[i:i+r])))
    stdTorque.append(getStdTorque)
    getRmsTorque = '%.6f'%float(abs(mean_squared_error(actualTorque[i:i+r],[0 for _ in actualTorque[i:i+r]], squared=False)))
    rmsTorque.append(getRmsTorque)


    mediaActivePower = '%.6f'%float(sum(actualActivePower[i:i+r])/len(actualActivePower[i:i+r]))
    avgActivePower.append(mediaActivePower) 
    getKurtosisActivePower = '%.6f'%float(abs(scipy.stats.kurtosis(actualActivePower[i:i+r],bias=False)))
    kurtosisActivePower.append(getKurtosisActivePower)
    getSkewnessActivePower = '%.6f'%float(abs(scipy.stats.skew(actualActivePower[i:i+r],bias= True)))
    skewnessActivePower.append(getSkewnessActivePower)
    getEntropyActivePower = '%.6f'%float(abs(scipy.stats.entropy(actualActivePower[i:i+r])))
    entropyActivePower.append(getEntropyActivePower)
    getMaxActivePower = '%.6f'%float(max(actualActivePower[i:i+r]))
    maxActivePower.append(getMaxActivePower)
    getMinActivePower = '%.6f'%float(min(actualActivePower[i:i+r]))
    minActivePower.append(getMinActivePower)
    getRangeActivePower = '%.6f'%float(float(getMaxActivePower)-float(getMinActivePower))
    rangeActivePower.append(getRangeActivePower)
    getStdActivePower = '%.6f'%float(abs(statistics.stdev(actualActivePower[i:i+r])))
    stdActivePower.append(getStdActivePower)
    getRmsActivePower = '%.6f'%float(abs(mean_squared_error(actualActivePower[i:i+r],[0 for _ in actualActivePower[i:i+r]], squared=False)))
    rmsActivePower.append(getRmsActivePower)

    
    
    perSecondSupport = int(i/r+1)
    perSecond.append(perSecondSupport)
    listClass.append(captureClass)

df = pd.DataFrame(zip(avgCurrent,stdCurrent,maxCurrent,minCurrent,rangeCurrent,kurtosisCurrent,rmsCurrent,skewnessCurrent,entropyCurrent,
                      avgTorque,stdTorque,maxTorque,minTorque,rangeTorque,kurtosisTorque,rmsTorque,skewnessTorque,entropyTorque,
                      avgActivePower,stdActivePower,maxActivePower,minActivePower,rangeActivePower,kurtosisActivePower,rmsActivePower,skewnessActivePower,entropyActivePower,listClass), 
                  columns=['AVG Current','STD Current', 'MAX Current', 'MIN Current', 'Range Current', 'Kurtosis Current', 'RMS Current', 'Skewness Current', 'Entropy Current', 'AVG Torque', 'STD Torque', 'MAX Torque', 'MIN Torque', 'Range Torque', 'Kurtosis Torque', 'RMS Torque', 'Skewness Torque', 'Entropy Torque','AVG Active Power', 'STD Active Power', 'MAX Active Power', 'MIN Active Power', 'Range Active Power', 'Kurtosis Active Power', 'RMS Active Power', 'Skewness Active Power', 'Entropy Active Power','Class'])
df.to_csv(fileName+'.csv',index=False)

files.download(fileName+'.csv')

fim = time.time()
print('tempo de execução: ',fim - inicio) 
```
